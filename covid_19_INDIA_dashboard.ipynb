```python
from bs4 import BeautifulSoup as bs
import pandas as pd
pd.set_option('display.max_colwidth', 500)
import time
import requests
import random
```

```python
page = requests.get("https://www.mohfw.gov.in/")
soup = bs(page.content)
```

```python
x = [i.text for i in soup.find_all(class_='col-xs-2')]
date = x[0][:58]
v = [i.text for i in soup.find_all(class_='coviddata')]
total_vaccination = (v[0][1:]).replace(',', '')
v_val = [i.text for i in soup.find_all(class_='coviddataval')]
vaccine_val = (v_val[0][1:10]).replace(',', '')
z = [i.text for i in soup.find_all(class_='mob-hide')]
active_cases = z[1][:7]
active_increase = z[1][13:18]
discharged = z[3][:8]
discharge_increase = z[3][14:20]
death = z[5][:6]
death_increase = z[5][15:19]
news = ([i.text for i in soup.find_all(class_='tested')][0]).replace(',','')
```

```python
print(f'_____________________________________________________________')
print()
print(f'{date}')
print(f'_____________________________________________________________')
print()
print(f'Active cases : {active_cases} ( {active_increase} ↑ )')
print(f'_____________________________________________________________')
print()
print(f'Discharged : {discharged} ( {discharge_increase} ↑ )')
print(f'_____________________________________________________________')
print()
print(f'Death : {death} ( {death_increase} ↑ )')
print(f'_____________________________________________________________')
print()
print(f'Total vaccination : {total_vaccination} ( {vaccine_val} ↑ )')
print(f'_____________________________________________________________')
print()
print(f'{news}')
print(f'_____________________________________________________________')
print()
```
